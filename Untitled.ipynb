```python
def clean_data(value):
    # Check if the input value is already a float
    if isinstance(value, float):
        return value
    
    # If not a float, proceed with cleaning
    value = value.split(' ')[0]
    value = value.strip()
    if value == "":
        value = 0
    return float(value)
```

```python
# Replace the column name
df.rename(columns={'mileage(km/ltr/kg)': 'mileage'}, inplace=True)
```

```python
# create our model

# Create the model formula
formula = 'selling_price ~ name + year + km_driven + fuel + seller_type + transmission + owner + mileage + engine + max_power + seats'
model1 = ols(formula=formula, data=df).fit()

# Print the summary of the model
model1.summary()
```

```python
# checking three assumptions of linear regression, we will visualize the normality of residuals 
import scipy.stats as stats
residuals = model1.resid
sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit = True)
plt.show()
```

```python
# to check homoscedasticity we will show in a graph the residuals and y predicted values
df1=df[['name', 'year', 'km_driven', 'fuel', 'seller_type', 'transmission', 'owner', 'mileage', 'engine', 'max_power', 'seats']].copy()
x=df1
y_pred = model1.predict(x)
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_pred, y=model1.resid,  data=df, color='red', alpha=0.7)
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residuals Plot for Homoscedasticity')
plt.show()
```

```python
# conclusion
# Following are the conclusions based on the first model we build

# R-squared and Adjusted R-squared are approximately same 0.617 and 0.616, which indicates that 61% of the variance 
# Prob(F-stats) means overall moel fit is 0.00, which means model fit is statistically significant
# p-values for most of the coefficients are 0.000 seem to be less than the significance level of 0.05 except name, owner, seats
# as we can see that from our model the most important variable is max_power
```

```python
# Conclusion

# In the model2 summary, the R-squared value is 0.769, which means that approximately 76.9% of the variability in the selling price is explained by the independent variables included in the model.
# Adj. R-squared: This is a version of R-squared that adjusts for the number of predictors in the model. It penalizes excessive use of predictors. In your model, the adjusted R-squared value is 0.767.
# t-statistic and P>|t|: The t-statistic measures the significance of each coefficient estimate. The associated p-value indicates the probability of observing the t-statistic or a more extreme value if the null hypothesis (the coefficient is zero) is true. Lower p-values indicate greater significance.
# [0.025 0.975]: These are the 95% confidence intervals for the corresponding coefficient estimates. They provide a range within which the true population parameter is likely to lie with 95% confidence.
# If P>|t| is less than 0.05, you can conclude that the coefficient is statistically significant.
# If P>|t| is greater than or equal to 0.05, you do not have enough evidence to reject the null hypothesis, and the coefficient is considered not statistically significant.

#  For example:

# The coefficient year has a p-value of 0.000, which is less than 0.05. Therefore, you would consider the coefficient for year to be statistically significant.
# On the other hand, the coefficient mileage has a p-value of 0.665, which is greater than 0.05. Therefore, you would not consider the coefficient for mileage to be statistically significant.
```

```python
# Creating  Model Three

continuous = ['year', 'selling_price','km_driven','mileage', 'engine', 'max_power']
```

```python
import numpy as np
import pandas as pd

# Define your DataFrame containing the continuous variables
# Assuming df_cont contains your continuous variables
df_cont = df[['year', 'selling_price', 'km_driven', 'mileage', 'engine', 'max_power']]

# Log Transformation
log_names = [f'{column}_log' for column in df_cont.columns]
df_log = np.log(df_cont.replace(0, np.nan))  # Replace 0 values with NaN to avoid undefined log(0)
df_log.columns = log_names

# Normalization
def normalize(feature):
    return (feature - feature.mean()) / feature.std()

df_log_norm = df_log.apply(normalize)
```

```python
# One Hot Encoding Categoricals

df_dummy = pd.get_dummies(df[categoricals], columns=['name', 'fuel', 'seller_type', 'transmission', 'owner', 'seats'], drop_first=True)

df_final = pd.concat([df_log_norm, df_dummy], axis=1)
```

```python
# Lets have a look at our final dataframe

df_final.head()
```

```python
# Checking Linearity between log dependent and independent continous varibales

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(16,3))

for xcol, ax in zip(['year_log','km_driven_log', 'mileage_log', 'engine_log', 'max_power_log'], axes):
    df_final.plot(kind='scatter', x=xcol, y='selling_price_log', ax=ax, alpha=0.4, color='b')

                         
```

```python
# Below we will check distribution of our dependent variable

sns.distplot(df_final['selling_price_log'])
```

```python
# create model 3

x = df_final.drop('selling_price_log', axis=1)
y = df_final['selling_price_log']
```

```python
# Impute missing values with the mean
x_filled = x.fillna(x.mean())

# Add a constant term for the intercept
x_int = sm.add_constant(x_filled)

# Fit the OLS regression model
model3 = sm.OLS(y.astype(float), x_int.astype(float)).fit()

# Print the summary of the model
print(model3.summary())
```

```python
model3.summary()
```

```python
# Checking normality of residuals in our model 3

import scipy.stats as stats
residuals = model3.resid
sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.show()
```

```python
# Checking homoscedasticity, we will show in a graph the residuals and y predictedvalues

y_pred3 = model3.predict(x_int)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_pred3, y=model3.resid,  data=df, color='red', alpha=0.7)
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residuals Plot for Homoscedasticity')
plt.show()
```

```python
# Conclusion

# The OLS regression results indicate that the model has an R-squared value of 0.896, indicating that approximately 89.6% of the variance in the target variable (selling_price_log) is explained by the independent variables included in the model.

# Some key points to note from the regression results are:

# The coefficients for the independent variables represent the change in the target variable for a one-unit change in the respective independent variable, holding all other variables constant.
# The p-values associated with each coefficient indicate the statistical significance of the corresponding independent variable. Variables with p-values less than the chosen significance level (usually 0.05) are considered statistically significant.
# The confidence intervals provide a range within which the true population parameter is likely to fall.
# The coefficients for year_log, km_driven_log, engine_log, and max_power_log are statistically significant at the 0.05 level, suggesting that these variables have a significant impact on the target variable.
# The coefficients for the categorical variables (e.g., name_2, fuel_2, seller_type_2, transmission_2, owner_2, etc.) represent the change in the target variable relative to the reference category (usually the category with the lowest index).
# Some categorical variables have coefficients with p-values less than 0.05, indicating that they are statistically significant predictors of the target variable.
# The Omnibus test, Durbin-Watson statistic, Jarque-Bera test, and Kurtosis provide information about the normality and autocorrelation of residuals.

# Based on these results, you can conclude that the model provides a good fit to the data and that several independent variables are statistically significant predictors of the target variable. However, further analysis and evaluation may be necessary to assess the model's predictive performance and identify any potential issues or areas for improvement.
```
